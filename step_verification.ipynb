# Energy Minimization

```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv('1-1_swp-cbd_trial1_em.xvg', comment='@', skiprows=12, names=['frame','energy'], delim_whitespace=True)
data1

plt.plot(data1.frame, data1.energy)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6,4), dpi=70)
plt.plot(data1.frame, data1.energy)
plt.title("GROMACS Energies \n 1-1 SWP-CBD Trial 1")
plt.legend(["Potential"])
plt.xlabel("Time (ps)")
plt.ylabel("Potential Energy (kJ/mol)")
plt.tight_layout()
fig.savefig("1-1_swp-cbd_trial1_em.jpg")
```
