```python
def singlePeptide(name):
    import pandas as pd
    import numpy as np
    import numpy.random
    import matplotlib.pyplot as plt
    
    fig = plt.figure(figsize=(10,6), dpi=70)
    
    for peptideNumber in range(1,6):
        xvgname = "{}_pep{}.xvg".format(name, peptideNumber)
        data = pd.read_csv(xvgname, comment='@', skiprows=13, names=['phi','psi','ram'], delim_whitespace=True)
        x = data.phi
        y = data.psi
        heatmap, xedges, yedges = np.histogram2d(x, y, bins=200)
        extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
        plt.subplot(2, 3, peptideNumber)
        plt.imshow(heatmap.T, extent=extent, origin='lower', cmap='inferno')
        plt.title("Peptide {}".format(peptideNumber))
        plt.tight_layout()
    
    fig.tight_layout()
    fig.savefig("{}.jpg".format(name))
```

```python
singlePeptide('1-1_swp-thc_trial1')
```
