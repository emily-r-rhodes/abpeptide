```python
def singlePeptide(name):
    import pandas as pd
    import numpy as np
    import numpy.random
    import matplotlib.pyplot as plt
    
    fig = plt.figure()
    
    for peptideNumber in range(1,6):
        xvgname = "{}_pep{}.xvg".format(name, peptideNumber)
        if peptideNumber == 1:
            data1 = pd.read_csv(xvgname, comment='@', skiprows=13, names=['phi','psi','ram'], delim_whitespace=True)
            x1 = data1.phi
            y1 = data1.psi
            heatmap1, xedges1, yedges1 = np.histogram2d(x1, y1, bins=200)
            extent1 = [xedges1[0], xedges1[-1], yedges1[0], yedges1[-1]]
            plt.clf()
            plt.imshow(heatmap1.T, extent=extent1, origin='lower', cmap='inferno')
            plt.show()
            #ax = fig.add_axes([0,0,1,1])
            #line = ax.plot(heatmap1.T, )
        if peptideNumber == 2:
            data2 = pd.read_csv(xvgname, comment='@', skiprows=13, names=['phi','psi','ram'], delim_whitespace=True)
            x2 = data2.phi
            y2 = data2.psi
            heatmap2, xedges2, yedges2 = np.histogram2d(x2, y2, bins=200)
            extent2 = [xedges2[0], xedges2[-1], yedges2[0], yedges2[-1]]
            plt.clf()
            plt.imshow(heatmap2.T, extent=extent2, origin='lower', cmap='inferno')
            plt.show()
        if peptideNumber == 3:
            data3 = pd.read_csv(xvgname, comment='@', skiprows=13, names=['phi','psi','ram'], delim_whitespace=True)
            x3 = data3.phi
            y3 = data3.psi
            heatmap3, xedges3, yedges3 = np.histogram2d(x3, y3, bins=200)
            extent3 = [xedges3[0], xedges3[-1], yedges3[0], yedges3[-1]]
            plt.clf()
            plt.imshow(heatmap3.T, extent=extent3, origin='lower', cmap='inferno')
            plt.show()
        if peptideNumber == 4:
            data4 = pd.read_csv(xvgname, comment='@', skiprows=13, names=['phi','psi','ram'], delim_whitespace=True)
            x4 = data4.phi
            y4 = data4.psi
            heatmap4, xedges4, yedges4 = np.histogram2d(x4, y4, bins=200)
            extent4 = [xedges4[0], xedges4[-1], yedges4[0], yedges4[-1]]
            plt.clf()
            plt.imshow(heatmap4.T, extent=extent4, origin='lower', cmap='inferno')
            plt.show()
        if peptideNumber == 5:
            data5 = pd.read_csv(xvgname, comment='@', skiprows=13, names=['phi','psi','ram'], delim_whitespace=True)
            x5 = data5.phi
            y5 = data5.psi
            heatmap5, xedges5, yedges5 = np.histogram2d(x5, y5, bins=200)
            extent5 = [xedges5[0], xedges5[-1], yedges5[0], yedges5[-1]]
            plt.clf()
            plt.imshow(heatmap5.T, extent=extent5, origin='lower', cmap='inferno')
            plt.show()
```
