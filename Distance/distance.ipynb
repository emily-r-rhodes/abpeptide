```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('1-1_swp-cbd_trial1_COLVAR', delim_whitespace=True)
```

```python
distance_values = list(data.columns.values)
distance_values.pop(0)
distance_values
```

```python
distance_values = list(data.columns.values)
distance_values.pop(0)

options = len(distance_values)
for n in range(0,options):
    plt.scatter(data.time, data[[distance_values[n]]], s=0.1)
    #plt.show() #show them all on different graphs
plt.show() #show them all on the same graph
```

```python
plt.scatter(data.time, data[[distance_values[n]]], s=0.1)
```

```python
distance_values = list(data.columns.values)
distance_values.pop(0)

options = len(distance_values)
for n in range(0,10):
    plt.plot(data.time, data[[distance_values[n]]], label=distance_values[n])
    #plt.show() #show them all on different graphs

plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1.5))
plt.show() #show them all on the same graph
```
